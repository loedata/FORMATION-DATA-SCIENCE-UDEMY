```python
# transformer les + et virgules pour faire du numériques (chiffres anglais)
# apply parcourt la colonne et pour chaque valeur elle transforme et passe au suivant
# fonction sur le vif plus stockée, temporaire
google_app_sans_nan['Installs']=google_app_sans_nan['Installs'].apply(lambda x: x.replace(',',''))
google_app_sans_nan['Installs']=google_app_sans_nan['Installs'].apply(lambda x: x.replace('+',''))
```

```python
# modifier colonne price pour en faire un qualitative (nombre)

# que contient prize de manière unique ==> fonction unique[]
google_app_sans_nan['Price'].unique()
```

```python
# prix avec dollars ==> suppression des $ par rien
google_app_sans_nan['Price']=google_app_sans_nan['Price'].apply(lambda x: x.replace("$",""))
# google_app_sans_nan['Price'].unique()

# transformer le prix en float
google_app_sans_nan['Price']=google_app_sans_nan['Price'].apply(lambda x: float(x))
```

```python
# catégorie ART_AND_DESIGN, FINANCE, BUSINESS... grouper selon colonne catégorie
google_app_sans_nan.groupby("Category")
```

```python
#L'objet est de type DataFrameGroupBy et plusieurs méthodes s'appliquent
# describe : seulement valeurs quantitative mais les stats sont faites par catégorie : regroupées
google_app_sans_nan.groupby("Category").describe()
```
