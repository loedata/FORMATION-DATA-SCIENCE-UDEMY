```python
#  0.8. sans NaN
google_play_store_sans_nan=google_play_store.dropna()
```

```python
# 1. Quels sont les noms, les catégories et les genres des applications 
# les plus installées sur Google Play?

# 1.1. Trouver le nombre d'installation le plus grand : fonction .max()
max_install=google_play_store_sans_nan['Installs'].max()
max_install
```

```python
# 1.2. Avec une condition, trouver la ligne qui correspond aux applis 
# avec les plus d'installation
# nom de la colonne pour les installations : Installs
# nom de la colonne pour le nom des applications : App
# nom de la colonne pour la catégorie : Category
# nom de la colonne pour le genre :Genres
google_play_store_sans_nan.loc[google_play_store_sans_nan['Installs'] == google_play_store_sans_nan['Installs'].max(),
['App','Category','Genres']]
```

```python
# 2.1. Quel est le plus grand nombre de Reviews (avis)?
max_avis=google_play_store_sans_nan['Reviews'].max()
print('Plus grand nombre d\'avis : %s' % max_avis)
```

```python
# 2.2. Quel est le nom et la note globale de l'application concernée?
google_play_store_sans_nan.loc[google_play_store_sans_nan['Reviews'] == max_avis,
['App','Rating']]
```

```python
# 3. Afficher les colonnes 2,5,6 et les lignes 3 à 16 du dataframe
# utiliser le slicing et index iloc
google_play_store_sans_nan.iloc[2:16,[1,4,5]]
```

```python
# 4. Combien y a-t-il d'applications ouvertes au personnes de tous âges (Everyone)?
# 7414 appli
google_play_store_sans_nan.loc[google_play_store_sans_nan['Content Rating'] == 'Everyone'].shape
```

```python
# 5. Dans quelle catégorie y a-t-il le plus d'applications?
# trouver la catégorie  où il y a le plus d'applications
# grouper par catégorie
# utiliser l'agrégation
google_play_store_sans_nan.groupby('Category').agg(['count'])
# d'après tableau ==> familly
```

```python
# 6. Quelle est l'application vendue la plus chère?
# 6.0. supprimer dollars et transformer en float
google_play_store_sans_nan['Price']=google_play_store_sans_nan['Price'].apply(lambda x: x.replace("$",""))
```

```python
# transformer le prix en float
google_play_store_sans_nan['Price']=google_play_store_sans_nan['Price'].apply(lambda x: float(x))
```

```python
# 6.1. trouver le plus de l'application la plus chère
prix_max=google_play_store_sans_nan['Price'].max()
print(prix_max)
```

```python
# 6.2. Afficher son nom
google_play_store_sans_nan.loc[google_play_store_sans_nan['Price'] == prix_max]
```
