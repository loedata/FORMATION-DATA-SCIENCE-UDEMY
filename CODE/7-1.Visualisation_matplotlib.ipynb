```python
# tracer la somme totale dépensées par chaque catégorie d'âge

# création d'un dictionnaire = structure adaptée pour tracer le graphique
age_groupe = black_friday.groupby('Age')
```

```python
# objet dataframegroupby
age_groupe
```

```python
# compter le nombre par age : fonction count()
age_groupe.count()
```

```python
# tracer un graphique de la somme des dépenses par catégorie d'âge

# création d'un dictionnaire vide
achats_client=dict()

# remplir le dictionnaire
for age, group in age_groupe:
    print('==========================================================')
    print(age)
    print('-----------------------------------------------------------')
    print(group)
    # groupe contient un sous dataframe contenant l'ensemble des lignes
    # de la catégorie d'âge l'une après l'autre
    # et comme on fait la somme on a toutes les dépenses
    achats_client[age]=sum(group['Purchase'])
```

```python
# les différentes types de graphique

# diagramme en baton

# ex : répartition par sexe du nombre d'achat

# création d'un jeu de donnée à partir de black friday
data=black_friday['Gender'].value_counts()
```

```python
# changer la couleur des barres : option color
# changer le paramètre 
plt.rcParams['figure.figsize']=[16,9]
# scatter créée des nuages de points automatiquement
plt.bar(data.index, data, color=['blue','pink'])
# dessine
plt.draw()
```

```python

# changer la couleur des barres : option color
# changer le paramètre 
plt.rcParams['figure.figsize']=[16,9]
# scatter créée des nuages de points automatiquement
plt.bar(data.index, data, color=['blue','pink'])
# donner un nom à l'axe des abscisses
plt.xlabel('Sexe de l''acheteur')
# donner un nom à l'axe des ordonnées
plt.ylabel('Nombre d''acheteurs')
# ajouter un titre enn remontant un peu le titre y= ou x=
plt.title('Nombre d''achats par sexe', y=1.05)
# dessine
plt.draw()
```

```python
# une fenêtre avec plusieurs graphiques : subplot sous graphiques


# 2 graphiques dans une fenâtre
plt.rcParams['figure.figsize']=[9,9]
# couper en 2 fois 1 moins graphique : 
# 2 lignes 1 colonnes et la première figure dans quel emplacement
# Premier graphique
plt.subplot(211)
plt.bar(data.index, data, color=['blue','pink'])
plt.xlabel('Sexe de l\'acheteur')
plt.ylabel('Nombre d\'acheteurs')
plt.title('Nombre d\'achats par sexe', y=1.05)
# Deuxième graphique
plt.subplot(212)
plt.boxplot(black_friday['Purchase'])
plt.xticks([1],['Ventes'])
plt.tight_layout()
plt.draw()
```
