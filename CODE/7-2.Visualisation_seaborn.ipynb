```python
# seaborn est une sous couche de matplotlib
# librairie plus puissante, plus facile que matplotlib

# importer la librairie pandas
import pandas

# importer le fichier pour le lire
black_friday=pandas.read_csv('BlackFriday.csv')

# importer la librairie seaborn
import seaborn as sns

# importer la librairie pyplot
import matplotlib.pyplot as plt

# retracer le boxplot mais par genre homme/femme
sns.boxplot(x='Gender',y='Purchase', data=black_friday, palette='Set1')
plt.draw()
```

```python
# 2 boplots un pour les femmes : un pour les hommes
# achats réalisés par les femmes et les hommes
# médiane de dépense des femmes plus petite que les hommes
# on n'utilise que les colonnes, pas la peine de préparer le dataset
```

```python
# palette de couleur cf site seaborn doc
sns.boxplot(x='Gender',y='Purchase', data=black_friday, palette='muted')
plt.draw()
```

```python
# selon quelle variable on colore les boxplot
sns.boxplot(x='Gender',y='Purchase', hue='Gender', data=black_friday,
            palette='muted')
plt.draw()
```

```python
# ajout 3 ième variables sur graphique, homme femme et catégorie age
sns.boxplot(x='Gender',y='Purchase', hue='Age', data=black_friday, palette='muted')
plt.draw()
```

```python
# placer la légende à l'extérieur du graphique
sns.boxplot(x='Gender',y='Purchase', hue='Age', data=black_friday, palette='muted')
# bbox_to_anchor : décaler à droite de 1,05 1
# loc=2 ==> légende en haut à gauche
# borderaxespar : décaler
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.draw()
```

```python
#  diagramme en bâton avec seaborn

# représenter le nombre de personnes par age et genre dans ce magasin
sns.countplot(x='Age', hue='Gender', data=black_friday, palette='Set1')
plt.draw()
```

```python
# scatterplot, séparé par catégorie : catplot
sns.catplot(x='Age',y='Purchase', data=black_friday)
plt.draw()
```

```python
# jeu de données aléatoires de 3000 lignes
subset=black_friday.sample(3000)
```

```python
subset
```

```python
sns.catplot(x='Age',y='Purchase', data=subset)
plt.draw()
```

```python
sns.catplot(x='Age',y='Purchase', data=subset, kind='swarm')
plt.draw()
```

```python
# colorer les points selon le genre
sns.catplot(x='Age',y='Purchase', hue='Gender', data=subset, kind='swarm')
plt.draw()
```

```python
# comment faire des subplot avec seaborn

# premier graphique
plt.subplot(211)
sns.countplot(x='Age', hue='Gender', data=black_friday, palette='Set1')

# second graphique
plt.subplot(212)
sns.boxplot(x='Gender',y='Purchase', hue='Age', data=black_friday, palette='muted')
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)

plt.draw()
```
